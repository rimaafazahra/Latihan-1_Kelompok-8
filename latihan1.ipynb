#SVD

```python
# Singular-value decomposition
from scipy.linalg import svd
# define a matrix
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
# SVD
U, s, VT = svd(A)
print(U)
print(s)
print(VT)
```

```python
# Reconstruct SVD
from numpy import dot

# define a matrix
A = np.array([[1, 2], [3, 4], [5, 6]])
print(A)
# Singular-value decomposition
U, s, VT = svd(A)
# create m x n Sigma matrix
Sigma = np.zeros((A.shape[0], A.shape[1])) # Sigma Tidak SQUARE
# populate Sigma with n x n diagonal matrix
Sigma[:A.shape[1], :A.shape[1]] = np.diag(s)
# reconstruct matrix
B = U.dot(Sigma.dot(VT))
print(B)
```

#**Case Study**
*The 20 Newsgroups data set is a collection of approximately 20,000 newsgroup documents, partitioned (nearly) evenly across 20 different newsgroups.*

Gunakan 20 Newsgroup data berikut ([Sumber Data](https://archive.ics.uci.edu/ml/datasets/Twenty+Newsgroups)). Bagaimana kita tahu topik apa yang dibicarakan dalam dokumen 20 Newsgroups tersebut? Mari lakukan implementasi *topic modelling* pada data tersebut.

```python
# Load libraries yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option("display.max_colwidth", 200)
```

```python
# Load data
from sklearn.datasets import fetch_20newsgroups

dataset = fetch_20newsgroups(shuffle=True, random_state=1, remove=('headers', 'footers', 'quotes'))
documents = dataset.data
```

```python
# Melihat dataset
documents[:2]
```

```python
print("Total Documents: ", len(documents))
```

Dataset ini memiliki **11.314** dokumen teks yang didistribusikan di **20** newsgroup yang berbeda.

```python
dataset.target_names
```

# Preprocessing

```python
# Muat data dari daftar ke dataframe untuk diproses lebih lanjut
news_df = pd.DataFrame({'document':documents})

# Menghapus semuanya kecuali huruf (yaitu tanda baca, angka, dan karakter khusus)
# dan akan menggantinya dengan ruang kosong
news_df['clean_doc'] = news_df['document'].str.replace("[^a-zA-Z#]", " ")

# menghapus kata-kata pendek karena biasanya tidak mengandung informasi yang berguna
news_df['clean_doc'] = news_df['clean_doc'].apply(lambda x: ' '.join([w for w in x.split() if len(w)>3]))

# buat semua teks menjadi huruf kecil untuk meniadakan sensitivitas huruf.
news_df['clean_doc'] = news_df['clean_doc'].apply(lambda x: x.lower())
```

Penting juga untuk menghapus stop-words dari data teks karena sebagian besar berantakan dan hampir tidak membawa informasi apa pun. Stop-words adalah istilah seperti 'it', 'they', 'am', 'been', 'about', 'because', 'while', 'the', dll.

Untuk menghapus stop-words dari dokumen, kita harus tokenize teksnya, yaitu, membagi string teks menjadi token atau kata-kata individual. Token tetap akan disatukan kembali setelah  menghapus stop-word.

```python
# Muat stopwords (di sini digunakan Spacy untuk stopwords, tapi juga dapat menggunakan NLTK)
from spacy.lang.en.stop_words import STOP_WORDS as stop_words

# tokenization - memecah kalimat menjadi kata-kata
tokenized_doc = news_df['clean_doc'].apply(lambda x: x.split())

# hapus stop-words
tokenized_doc = tokenized_doc.apply(lambda x: [item for item in x if item not in stop_words])

# de-tokenization - Gabungkan kata-kata untuk mendapatkan kembali kalimatnya
detokenized_doc = []
for i in range(len(news_df)):
    t = ' '.join(tokenized_doc[i])
    detokenized_doc.append(t)

news_df['clean_doc'] = detokenized_doc
```

Ini adalah langkah pertama menuju pemodelan topik. Kami akan menggunakan TfidfVectorizer sklearn untuk membuat matriks istilah dokumen dengan **1.000** istilah.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(stop_words='english', 
max_features= 1000, # keep top 1000 terms 
max_df = 0.5, 
smooth_idf=True)

X = vectorizer.fit_transform(news_df['clean_doc'])

X.shape # check shape of the document-term matrix
```

Jadi ***X*** adalah matriks yang sama dengan **U** yang ditunjukkan di atas.

CATATAN: Karena batasan komputasi,  jumlah fitur telah dibatasi hingga 1.000.

*   Setiap baris dari matriks ***Uk*** (document-term matrix) adalah representasi vektor dari dokumen yang bersangkutan. Panjang vektor ini adalah ***k***, yang merupakan jumlah topik yang diinginkan. Representasi vektor untuk suku-suku dalam data kami dapat ditemukan dalam matriks ***Vk*** (matriks istilah-topik).
*   Jadi, SVD memberi kita vektor untuk setiap dokumen dan istilah dalam data kita. Panjang setiap vektor adalah ***k***. Kami kemudian dapat menggunakan vektor-vektor ini untuk menemukan kata-kata dan dokumen serupa menggunakan metode kesamaan kosinus.

# Topic Modelling

Langkah selanjutnya adalah merepresentasikan setiap istilah dan dokumen sebagai vektor. Kami akan menggunakan matriks istilah dokumen dan menguraikannya menjadi beberapa matriks. Kami akan menggunakan sklearn's TruncatedSVD untuk melakukan tugas dekomposisi matriks.

Karena data berasal dari 20 newsgroup yang berbeda, mari kita coba memiliki 20 topik untuk data teks kita. Jumlah topik dapat ditentukan dengan menggunakan parameter n_components.

```python
from sklearn.decomposition import TruncatedSVD

# SVD represent documents and terms in vectors 
svd_model = TruncatedSVD(n_components=20, algorithm='randomized', n_iter=100, random_state=122)

svd_model.fit(X)

len(svd_model.components_)
```

Komponen svd_model adalah topiknya, dan dapat diakses menggunakan ***svdmodel.components***. Terakhir, mari cetak beberapa kata terpenting di masing-masing dari ***20*** topik dan lihat bagaimana model melakukannya.

```python
terms = vectorizer.get_feature_names()
for i, comp in enumerate(svd_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key= lambda x:x[1], reverse=True)[:7]
    print("Topic "+str(i)+": ", end='', flush=True)
    for t in sorted_terms:
        print(t[0], end=' ', flush=True)
    print(" ")
```

**Kata-kata yang ditunjukkan di atas merupakan informasi yang paling sering muncul pada tiap artikel/topic.**

Untuk mengetahui seberapa berbeda topik kita, kita harus memvisualisasikannya. Tentu saja, kita tidak dapat memvisualisasikan lebih dari 3 dimensi, tetapi ada teknik seperti ***PCA*** dan ***t-SNE*** yang dapat membantu kita memvisualisasikan data berdimensi tinggi ke dalam dimensi yang lebih rendah. Disini kita akan menggunakan ***t-SNE***.

```python
X_topics = svd_model.fit_transform(X)
```

```python
from sklearn.manifold import TSNE
embedding = TSNE(n_components=2).fit_transform(X_topics)
```

```python
plt.figure(figsize=(7,5))
plt.scatter(embedding[:, 0], embedding[:, 1],
            c = dataset.target,
            s = 10, # size
            edgecolor='none')
plt.show()
```

**Plot di atas menunjukkan persebaran kata-kata dari setiap topic, di mana masing-masing topic dibedakan dengan warna.**

Seperti yang Anda lihat di atas, hasilnya cukup bagus. Setiap titik mewakili dokumen dan warna mewakili 20 newsgroup.


```python
# melakukan import module WordCloud untuk membuat wordcloud

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# membuat list kosong yang akan digunakan untuk menyimpan data

topic_terms_table=[]
topic_terms_pertopic=[]

# membuat for loop untuk menyimpan data berupa nomor topic, topic, dan frequency

terms = vectorizer.get_feature_names()
for i, comp in enumerate(svd_model.components_):
    terms_comp = zip(terms, comp)
    sorted_terms = sorted(terms_comp, key= lambda x:x[1], reverse=True)
    for t in sorted_terms:
        topic_terms_pertopic = [i,t[0],t[1]]
        topic_terms_table.append(topic_terms_pertopic)
```

```python
# membuat data frame panda untuk memudahkan pengambilan data sesuai topic

topic_terms_table_pd = pd.DataFrame(topic_terms_table)
topic_terms_table_pd.head()
```

```python
# membuat fungsi yang dapat melakukan return data sesuai topic yang diinginkan

def frequencies_data(notopics):
    for i in range(len(topic_terms_table_pd)):
        pertopic_table = topic_terms_table_pd[topic_terms_table_pd[0] == notopics]
    pertopic_table = pertopic_table.drop(0, axis = 1)
    pertopic_tuples = list(pertopic_table.itertuples(index=False, name=None))
    return pertopic_tuples
```

```python
wordcloud = WordCloud(min_word_length =3,
                      background_color='white')

# men-generate word cloud yang ingin dibuat
for i in range(20):
    wordcloud.generate_from_frequencies(dict(frequencies_data(i)))

    # melakukan print word cloud
    print("Word Cloud Topic "+str(i))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
    print('')
```

**Berikut adalah visualisasi berupa Word Cloud yang juga menunjukkan informasi-informasi yang paling sering muncul pada tiap topic.**

```python
def bar_topic(notopics):
    for i in range(len(topic_terms_table_pd)):
        pertopic_bar_table = topic_terms_table_pd[topic_terms_table_pd[0] == notopics]
    pertopic_bar_table = pertopic_bar_table.drop(0, axis = 1)
    return pertopic_bar_table.head(10)
```

```python
for i in range(20):
    # Figure Size
    fig, ax = plt.subplots(figsize =(16, 9))
    # Horizontal Bar Plot
    ax.barh(bar_topic(i)[1], bar_topic(i)[2])
    # Remove axes splines
    for s in ['top', 'bottom', 'left', 'right']:
        ax.spines[s].set_visible(False)
    
    # Remove x, y Ticks
    ax.xaxis.set_ticks_position('none')
    ax.yaxis.set_ticks_position('none')
    
    # Add padding between axes and labels
    ax.xaxis.set_tick_params(pad = 5)
    ax.yaxis.set_tick_params(pad = 10)
    
    # Add x, y gridlines
    ax.grid(b = True, color ='grey',
            linestyle ='-.', linewidth = 0.5,
            alpha = 0.2)
    
    # Show top values
    ax.invert_yaxis()
    
    # Add annotation to bars
    for a in ax.patches:
        plt.text(a.get_width()+0.2, a.get_y()+0.5,
                str(round((a.get_width()), 2)),
                fontsize = 10, fontweight ='bold',
                color ='grey')
    
    # Add Plot Title
    ax.set_title("Top 10 Words In Topic "+str(i),
                 loc ='left')
    
    # Show Plot
    plt.show()
```

**Tampilan di atas merupakan bentuk Bar Plot dari 10 kata yang sering muncul di tiap topic.**
